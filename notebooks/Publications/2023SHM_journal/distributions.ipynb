```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-10, 10, 0.001)
# Mean = 0, SD = 2.
plt.plot(x_axis, norm.pdf(x_axis,0,3), color='black', linestyle='-', linewidth=10)


# Remove ticks
plt.xticks([])
plt.yticks([])
plt.xlabel('$f_{res}$', fontsize=100)
plt.box(False)
plt.show()

# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-10, 10, 0.001)
# Mean = 0, SD = 2.
plt.plot(x_axis, norm.pdf(x_axis,0,1), color='black', linestyle='-', linewidth=10)


# Remove ticks
plt.xticks([])
plt.yticks([])
plt.xlabel('$f_{res}$', fontsize=100)
plt.box(False)
plt.show()
```
